````python
import pandas as pd
import json
````

````python
with open('points010220240920.json', 'r') as file:
    points_data = json.load(file)

points_df = pd.DataFrame(points_data.items(), columns=['address', 'balance'])
points_df['balance'] = points_df['balance'].apply(lambda x: int(x))
print("Surge original data")
points_df.describe()
````

````python
print("Group by balances")
points_df.groupby('balance').size().reset_index(name='count').sort_values('balance', ascending=True).head(30)
````

````python
with open('points010220240920_anonymous.json', 'r') as file:
    points_anon_data = json.load(file)

points_anon_df = pd.DataFrame(points_anon_data.items(), columns=['address', 'balance'])
points_anon_df['balance'] = points_anon_df['balance'].apply(lambda x: int(x))
print("Original data")
points_anon_df.describe()
````

````python
print("Number of address for the lowest balances groups")
points_anon_df.groupby('balance').size().reset_index(name='count').sort_values('balance', ascending=True).head(20)
````

````python
print("Number of address for the hightest balances groups")
points_anon_df.groupby('balance').size().reset_index(name='count').sort_values('balance', ascending=False).head(20)
````

````python
print("Balances with more address grouped")
points_anon_df.groupby(["balance"]).count().sort_values('address', ascending=False).head(20)
````

````python
print("Balances with less address grouped")
points_anon_df.groupby(["balance"]).count().sort_values('address', ascending=True).head(20)
````
